```python
#drop null value of ingredients,dry,oily
df.dropna(subset = ['Ingredients','Dry','Oily'],inplace=True)
df
```

```python
#which brands have most ranks
print(df['Brand'].value_counts())
plt.figure(figsize = (50,50))
sns.countplot(data = df, x = 'Brand',order = df['Brand'].value_counts().index, color = 'green')
plt.xlabel('Brands',fontsize = 40)
plt.ylabel('count', fontsize = 40)
plt.xticks(rotation = 90)
plt.show()
```

```python
# take all products and show the higest demainding product
print(df['Label'].value_counts())
plt.figure(figsize=(10,10))
sns.countplot(data = df, x ='Label', order = df['Label'].value_counts().index, color = 'orange')
plt.xlabel('Label', fontsize = 30)
plt.ylabel('count', fontsize = 30)
plt.xticks(rotation = 90)
plt.show()
```
